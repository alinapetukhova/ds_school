```python
%matplotlib inline
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")

from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
data = pd.read_csv('data_ohe.csv')
y = pd.read_csv('data.csv').target
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.3, random_state=1)
```

```python
abl = AdaBoostClassifier()
abl.fit(X=X_train, y=y_train)
accuracy_score(y_test, abl.predict(X_test))
```

```python
gbm = XGBClassifier()
gbm.fit(X=X_train, y=y_train)
accuracy_score(y_test, gbm.predict(X_test))
```

```python
lgbc =  LGBMClassifier()
lgbc.fit(X=X_train, y=y_train)
accuracy_score(y_test, lgbc.predict(X_test))
```

```python
cbc = CatBoostClassifier()
cbc.fit(X=X_train, y=y_train)
accuracy_score(y_test, cbc.predict(X_test))
```

```python
# Create ensemble
mean_errors = (abl.predict_proba(X_test) + \
               gbm.predict_proba(X_test) + \
               lgbc.predict_proba(X_test) + \
               cbc.predict_proba(X_test)) / 4
```

```python
final_pred = [1 if i[1]>0.5 else 0 for i in mean_errors]
accuracy_score(y_test, final_pred)
```

```python
names = ['AdaBoost', 'XGBoost', 'LGBoost', 'CatBoost', 'Ensemble']
values = [accuracy_score(y_test, abl.predict(X_test)), accuracy_score(y_test, gbm.predict(X_test)),
          accuracy_score(y_test, lgbc.predict(X_test)), accuracy_score(y_test, cbc.predict(X_test)),
          accuracy_score(y_test, final_pred)]

plt.figure(1, figsize=(9, 3))

plt.barh(names, values, )
plt.suptitle('Accuracy Plotting')
plt.xlim(0.8, 0.9)
plt.show()
```
